```python
import heapq

GOAL = ((1,2,3),
        (4,5,6),
        (7,8,0))

DIRS = [(-1,0),(1,0),(0,-1),(0,1)]


def manhattan(state):
    dist = 0
    for i in range(3):
        for j in range(3):
            val = state[i][j]
            if val != 0:
                gx = (val-1)//3
                gy = (val-1)%3
                dist += abs(i - gx) + abs(j - gy)
    return dist


def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j


def neighbors(state):
    x, y = find_blank(state)
    res = []
    for dx, dy in DIRS:
        nx, ny = x+dx, y+dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new = [list(r) for r in state]
            new[x][y], new[nx][ny] = new[nx][ny], new[x][y]
            res.append(tuple(tuple(r) for r in new))
    return res


def is_solvable(state):
    flat = [n for row in state for n in row if n != 0]
    inv = 0
    for i in range(len(flat)):
        for j in range(i+1, len(flat)):
            if flat[i] > flat[j]:
                inv += 1
    return inv % 2 == 0


def print_board(state):
    for row in state:
        print(" ".join(str(x) if x != 0 else "_" for x in row))
    print()


def astar(start):
    pq = []
    heapq.heappush(pq, (manhattan(start), 0, start, []))
    visited = set()

    while pq:
        f, g, state, path = heapq.heappop(pq)

        if state in visited:
            continue
        visited.add(state)

        new_path = path + [state]

        if state == GOAL:
            return new_path

        for nb in neighbors(state):
            if nb not in visited:
                heapq.heappush(
                    pq,
                    (g + 1 + manhattan(nb), g + 1, nb, new_path)
                )

    return None

print("Enter puzzle row by row (use 0 for blank):")

grid = []
for i in range(3):
    row = list(map(int, input().split()))
    grid.append(tuple(row))

start = tuple(grid)

if not is_solvable(start):
    print("\nThis puzzle has NO solution.")
else:
    solution = astar(start)

    for i, step in enumerate(solution):
        print(f"Step {i}:")
        print_board(step)

    print("Total moves:", len(solution)-1)
```
