```python
from collections import deque


def add_edge(graph, source, destination):
    """Add an undirected edge to the graph."""
    graph[source].append(destination)
    graph[destination].append(source)


def dfs_recursive(vertex, graph, visited, order):
    """Perform DFS recursively from the given vertex."""
    visited[vertex] = True
    order.append(vertex)
    
    for neighbor in graph[vertex]:
        if not visited[neighbor]:
            dfs_recursive(neighbor, graph, visited, order)


def dfs_traversal_from_start(vertices, graph, start_vertex):
    """Perform DFS traversal starting from start_vertex."""
    visited = [False] * vertices
    order = []
    
    dfs_recursive(start_vertex, graph, visited, order)
    
    for vertex in range(vertices):
        if not visited[vertex]:
            dfs_recursive(vertex, graph, visited, order)
    
    return order


def bfs_traversal_from_start(vertices, graph, start_vertex):
    """Perform BFS traversal starting from start_vertex."""
    visited = [False] * vertices
    order = []
    queue = deque()
    
    visited[start_vertex] = True
    queue.append(start_vertex)
    
    while queue:
        current = queue.popleft()
        order.append(current)
        
        for neighbor in graph[current]:
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)
    
    for vertex in range(vertices):
        if not visited[vertex]:
            visited[vertex] = True
            queue.append(vertex)
            
            while queue:
                current = queue.popleft()
                order.append(current)
                
                for neighbor in graph[current]:
                    if not visited[neighbor]:
                        visited[neighbor] = True
                        queue.append(neighbor)
    
    return order


def main():
    """Main function to read input and perform graph traversals."""
    # Read number of vertices
    vertices = int(input("Enter number of vertices: "))
    if vertices <= 0:
        print("Number of vertices must be greater than 0.")
        return
    
    graph = [[] for _ in range(vertices)]
    
    edges = int(input("Enter number of edges: "))
    
    print("Enter edges (u v):")
    for _ in range(edges):
        edge_input = input().split()
        source = int(edge_input[0])
        destination = int(edge_input[1])
        
        if source < 0 or source >= vertices or destination < 0 or destination >= vertices:
            print(f"Invalid edge: {source} {destination}")
            return
        
        add_edge(graph, source, destination)
    
    for vertex in range(vertices):
        graph[vertex].sort()
    
    start_vertex = int(input("Enter starting vertex: "))
    if start_vertex < 0 or start_vertex >= vertices:
        print("Invalid starting vertex.")
        return
    
    dfs_order = dfs_traversal_from_start(vertices, graph, start_vertex)
    bfs_order = bfs_traversal_from_start(vertices, graph, start_vertex)
    
    print(f"DFS visit order: {dfs_order}")
    print(f"BFS visit order: {bfs_order}")


main()
```
