## Branch and Bound and Backtracking for n-queens problem

```python
import sys

def normalize_input(r, c, N):
    one_based = (r >= 1 and r <= N and c >= 1 and c <= N)
    zero_based = (r >= 0 and r < N and c >= 0 and c < N)
    if one_based and not zero_based:
        return r - 1, c - 1
    if zero_based:
        return r, c
    return None

def main():
    try:
        N = int(input("What is the grid size? "))
        coords_str = input("Enter first queen coordinate (row col): ")
        coords = coords_str.strip().split()
        r, c = int(coords[0]), int(coords[1])
    except ValueError:
        print("Invalid input.")
        return

    norm = normalize_input(r, c, N)
    if norm is None:
        print(f"Invalid coordinates. Use 0..{N-1} or 1..{N} consistently.")
        return
    
    r, c = norm

    col_for_row = [-1] * N
    cols_used = [False] * N
    diag1_used = [False] * (2 * N)
    diag2_used = [False] * (2 * N)
    fixed_row = r
    fixed_col = c

    def place_fixed_queen(r_idx, c_idx):
        if r_idx < 0 or r_idx >= N or c_idx < 0 or c_idx >= N:
            return False
        col_for_row[r_idx] = c_idx
        cols_used[c_idx] = True
        diag1_used[r_idx - c_idx + N] = True
        diag2_used[r_idx + c_idx] = True
        return True

    def solve(row):
        if row >= N:
            return True
        if row == fixed_row:
            return solve(row + 1)
            
        for col in range(N):
            if cols_used[col]:
                continue
            d1 = row - col + N
            d2 = row + col
            if diag1_used[d1] or diag2_used[d2]:
                continue
                
            col_for_row[row] = col
            cols_used[col] = True
            diag1_used[d1] = True
            diag2_used[d2] = True
            
            if solve(row + 1):
                return True
                
            col_for_row[row] = -1
            cols_used[col] = False
            diag1_used[d1] = False
            diag2_used[d2] = False
            
        return False

    def print_board_with_attacks(positions, show_attacks):
        board = [['.' for _ in range(N)] for _ in range(N)]
        
        for row in range(N):
            col = positions[row]
            if col == -1:
                continue
            board[row][col] = 'Q'
            
        attacked = [[False for _ in range(N)] for _ in range(N)]
        for row in range(N):
            col = positions[row]
            if col == -1:
                continue
            for j in range(N):
                if j != col:
                    attacked[row][j] = True
            for i in range(N):
                if i != row:
                    attacked[i][col] = True
                    
            i, j = row - 1, col - 1
            while i >= 0 and j >= 0:
                attacked[i][j] = True
                i -= 1; j -= 1
            i, j = row - 1, col + 1
            while i >= 0 and j < N:
                attacked[i][j] = True
                i -= 1; j += 1
            i, j = row + 1, col - 1
            while i < N and j >= 0:
                attacked[i][j] = True
                i += 1; j -= 1
            i, j = row + 1, col + 1
            while i < N and j < N:
                attacked[i][j] = True
                i += 1; j += 1

        for i in range(N):
            for j in range(N):
                ch = board[i][j]
                if ch == 'Q':
                    sys.stdout.write("Q ")
                elif show_attacks and attacked[i][j]:
                    sys.stdout.write("x ")
                else:
                    sys.stdout.write(". ")
            sys.stdout.write("\n")

    if not place_fixed_queen(r, c):
        print("Fixed queen position is invalid (out of range).")
        return
        
    print("Initial board with the fixed queen and its attack path:")
    print_board_with_attacks(col_for_row, True)
    
    solved = solve(0)
    
    if solved:
        print("\nSolution found:")
        print_board_with_attacks(col_for_row, False)
    else:
        print(f"\nNo solution possible with the fixed first queen at ({fixed_row + 1},{fixed_col + 1}).")

if __name__ == '__main__':
    main()
```
