```python
# 1) Greedy Selection Sort
```

```python
def find_minimum_index(array, start_position):
    minimum_index = start_position
    for current_index in range(start_position + 1, len(array)):
        if array[current_index] < array[minimum_index]:
            minimum_index = current_index
    return minimum_index

def swap(array, first_index, second_index):
    array[first_index], array[second_index] = array[second_index], array[first_index]

def selection_sort(array):
    for current_position in range(len(array) - 1):
        minimum_index = find_minimum_index(array, current_position)
        swap(array, current_position, minimum_index)

def main():
    array_length = int(input("Enter array length: "))
    array = []
    print(f"Enter {array_length} elements in one line:")
    input_line = input().strip()
    input_elements = input_line.split() if input_line else []

    current_index = 0
    for elem in input_elements:
        if current_index < array_length:
            array.append(int(elem))
            current_index += 1

    while current_index < array_length:
        try:
            value = int(input())
            array.append(value)
            current_index += 1
        except ValueError:
            break

    print("Before:", array)
    selection_sort(array)
    print("After: ", array)

main()
```

```python
# 2) Greedy search algorithm for Prim's Minimal Spanning Tree Algorithm
```

```python

import heapq

class Edge:
    def __init__(self, source, destination, weight):
        self.source = source
        self.destination = destination
        self.weight = weight
    def __lt__(self, other):
        return self.weight < other.weight

def prims_algorithm(num_vertices, adj, one_based_input):
    visited = [False] * num_vertices
    pq = []
    mst = []
    min_cost = 0

    visited[0] = True
    for edge in adj[0]:
        heapq.heappush(pq, edge)

    while pq and len(mst) < num_vertices - 1:
        current_edge = heapq.heappop(pq)
        u = current_edge.source
        v = current_edge.destination
        weight = current_edge.weight

        if visited[u] and visited[v]:
            continue

        if not visited[v]:
            mst.append(current_edge)
            min_cost += weight
            visited[v] = True
            for edge in adj[v]:
                if not visited[edge.destination]:
                    heapq.heappush(pq, edge)
        elif not visited[u]:
            mst.append(current_edge)
            min_cost += weight
            visited[u] = True
            for edge in adj[u]:
                if not visited[edge.destination]:
                    heapq.heappush(pq, edge)

    print("Edges in the Minimum Spanning Tree:")
    print_offset = 1 if one_based_input else 0
    for edge in mst:
        print(f"{edge.source + print_offset} - {edge.destination + print_offset} (Weight: {edge.weight})")
    print(f"Minimum Spanning Tree Cost: {min_cost}")

    if len(mst) != num_vertices - 1:
        print("Warning: Graph may be disconnected. MST is incomplete.")

def main():
    num_vertices = int(input("Enter number of vertices: "))
    num_edges = int(input("Enter number of edges: "))

    adj = [[] for _ in range(num_vertices)]
    print("Enter edges (u v cost):")
    input_u = []
    input_v = []
    input_cost = []
    min_vertex = float('inf')
    max_vertex = float('-inf')

    for _ in range(num_edges):
        u, v, cost = map(int, input().split())
        input_u.append(u)
        input_v.append(v)
        input_cost.append(cost)
        min_vertex = min(min_vertex, u, v)
        max_vertex = max(max_vertex, u, v)

    is_zero_based = (min_vertex == 0 and max_vertex <= num_vertices - 1)
    is_one_based = (min_vertex >= 1 and max_vertex <= num_vertices)

    if not is_zero_based and not is_one_based:
        print("Invalid vertex labels.")
        print(f"Use either 0 to {num_vertices - 1} OR 1 to {num_vertices} consistently.")
        return

    for i in range(num_edges):
        normalized_u = input_u[i] - 1 if is_one_based else input_u[i]
        normalized_v = input_v[i] - 1 if is_one_based else input_v[i]
        cost = input_cost[i]
        adj[normalized_u].append(Edge(normalized_u, normalized_v, cost))
        adj[normalized_v].append(Edge(normalized_v, normalized_u, cost))

    prims_algorithm(num_vertices, adj, is_one_based)


main()
```
