## Descriptive Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

## Grouped Summary Statistics

```python
stats = df.groupby("Species")["SepalLengthCm"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Std_Deviation="std"
)

print(stats)
```

## BAR CHART – Mean Values

```python
plt.figure()
stats["Mean"].plot(kind="bar")
plt.title("Mean Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Mean Sepal Length (cm)")
plt.show()
```

## BOXPLOT – Variability

```python
plt.figure()
df.boxplot(column="SepalLengthCm", by="Species")
plt.title("Sepal Length Distribution by Species")
plt.suptitle("")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

## Percentiles and Stats

```python
for sp in df["Species"].unique():
    print("\nStatistics for", sp)
    print(df[df["Species"]==sp][
        ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]
    ].describe())
```
