```python
## Title Data Visualization II
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

#titanic dataset
data = pd.read_csv("titanic_train.csv")

#tips dataset
tips = sns.load_dataset("tips")
```

```python
## Pie Chart of Sex
```

```python
data['Sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
## Histogram of Age
```

```python
plt.hist(data['Age'], bins=5)
plt.show()
```

```python
## Distplot of Age
```

```python
sns.histplot(data['Age'])
plt.show()
```

```python
## Scatter Plot of Total Bill vs Tip
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
## Multivariate Scatter Plot (Total Bill, Tip, Sex)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.show()
```

```python
## Multivariate Scatter Plot (Total Bill, Tip, Sex, Smoker)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips)
plt.show()
```

```python
## Bar Plot of Pclass vs Age
```

```python
sns.barplot(x="Pclass", y="Age", data=data)
plt.show()
```

```python
## Multivariate Bar Plot of Pclass, Fare, and Sex
```

```python
sns.barplot(x="Pclass", y="Fare", hue="Sex", data=data)
plt.show()
```

```python
## Boxplot of Sex vs Age
```

```python
sns.boxplot(x="Sex", y="Age", data=data)
plt.show()
```

```python
## Multivariate Boxplot of Sex, Age, and Survived
```

```python
sns.boxplot(x="Sex", y="Age", hue="Survived", data=data)
plt.show()
```

```python
## Distplot of Age by Survival
```

```python
sns.kdeplot(data=data[data['Survived']==0]['Age'], label="Not Survived")
sns.kdeplot(data=data[data['Survived']==1]['Age'], label="Survived")

plt.legend()
plt.show()
```

```python
## Crosstab of Pclass and Survived
```

```python
pd.crosstab(data['Pclass'], data['Survived'])
```

```python
## Heatmap of Pclass and Survived
```

```python
sns.heatmap(pd.crosstab(data['Pclass'], data['Survived']))
```

```python
## Cluster Map of Parch and Survived
```

```python
sns.clustermap(pd.crosstab(data['Parch'], data['Survived']))
plt.show()
```
