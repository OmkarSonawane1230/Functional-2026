```python
## Data Visualization I
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
## Distribution Plot of Ticket Prices (Fare)
```

```python
sns.histplot(dataset['fare'])
sns.histplot(dataset['fare'], kde=False, bins=10)
```

```python
## Joint Plot of Age vs Fare
```

```python
sns.jointplot(x='age', y='fare', data=dataset)
sns.jointplot(x='age', y='fare', data=dataset, kind='hex')
```

```python
## Pair Plot for Dataset Features
```

```python
dataset = dataset.dropna()
sns.pairplot(dataset)
sns.pairplot(dataset, hue='sex')
```

```python
## Rug Plot of Ticket Prices
```

```python
sns.rugplot(dataset['fare'])
```

```python
## Bar Plot of Average Age by Gender
```

```python
sns.barplot(x='sex', y='age', data=dataset)
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
```

```python
## Count Plot of Passengers by Gender
```

```python
sns.countplot(x='sex', data=dataset)
```

```python
## Box Plot of Age by Gender and Survival
```

```python
sns.boxplot(x='sex', y='age', data=dataset)
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

```python
## Violin Plot of Age Distribution by Gender and Survival
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
sns.violinplot(x='sex', y='age', data=dataset, hue='survived', split=True)
```

```python
## Strip Plot of Age by Gender and Survival
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived', dodge=True)
```

```python
## Swarm Plot of Age by Gender and Survival
```

```python
sns.swarmplot(x='sex', y='age', data=dataset)
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
## Combined Swarm and Violin Plot
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
sns.swarmplot(x='sex', y='age', data=dataset, color='black')
```
