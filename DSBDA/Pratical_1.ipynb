```python
import pandas as pd
import numpy as np
```

## Load Dataset into Pandas DataFrame

```python
df = pd.read_csv("train.csv")
df.head()
```

## Data Preprocessing & Initial Exploration

```python
df.shape
df.info()
```

```python
df.describe()
df.isnull().sum()
```

## Data Formatting & Normalization

```python
df.dtypes
df['Survived'] = df['Survived'].astype('category')
```

## Convert Categorical → Quantitative Variables

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
```

```python
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
df['Cabin'] = df['Cabin'].apply(lambda x: 0 if x == 'Unknown' else 1)
```

```python
df[['Embarked_C', 'Embarked_Q', 'Embarked_S']] = df[['Embarked_C', 'Embarked_Q', 'Embarked_S']].astype(int)
```
